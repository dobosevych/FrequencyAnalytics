```python
import nltk
import string
import py

from bs4 import BeautifulSoup
from collections import Counter
from matplotlib import pyplot as plt
```

```python
html_doc = open('../data/zakhar-berkut.html').read()
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
pars = soup.find_all('p')
```

```python
nltk.word_tokenize(pars[0].get_text().lower())
```

```python
punctuation = string.punctuation + '—'
punctuation
```

```python
all_words = []
for par in pars:
    all_words += [word for word in nltk.word_tokenize(par.get_text().lower()) if word not in punctuation]
```

```python
all_words[12]
```

```python
count = Counter(all_words)
count.most_common(15)
```

```python
words = count.most_common(25)

fig = plt.figure(figsize=(10,10), dpi=100)
ax = fig.add_subplot(111)

plt.plot([pair[1] for i, pair in enumerate(words)])
          
for i, pair in enumerate(words):
    ax.annotate(pair[0], (i, pair[1]))
plt.plot()
plt.show()
```
